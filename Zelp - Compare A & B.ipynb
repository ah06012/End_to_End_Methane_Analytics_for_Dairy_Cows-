### Comparing Gas data - Cow A vs Cow B

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np

# 3x2 grid of subplots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(16, 15), gridspec_kw={'height_ratios': [1, 1, 1]})

# ======================
# TOP ROW: Cow A Gas Concentration
# ======================

# TOP LEFT: Cow A - CO2 & CO2Temp
ax1_left = axes[0, 0]
ax1_right = ax1_left.twinx()

# Plot CO2
ax1_left.plot(df_cow_A_gas['timestamp'], df_cow_A_gas['co2'], label='CO2 (Concentration)', color='blue')
ax1_left.set_xlabel('Time')
ax1_left.set_ylabel('CO2 Concentration (%)')
ax1_left.set_title('Cow A: CO2 & CO2Temp Over Time - 13/08/2024')

# Plot CO2Temp
ax1_right.scatter(df_cow_A_gas['timestamp'], df_cow_A_gas['co2temp'], label='CO2 Temp', 
                 color='green', s=10, alpha=0.7)
ax1_right.set_ylabel('CO2 Temperature (°C)')

# x-axis show only HH:MM 
ax1_left.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
ax1_left.xaxis.set_major_locator(mdates.MinuteLocator(byminute=range(0, 60, 10)))

# legends
lines_left, labels_left = ax1_left.get_legend_handles_labels()
lines_right, labels_right = ax1_right.get_legend_handles_labels()
ax1_right.legend(lines_left + lines_right, labels_left + labels_right, loc='upper left')

# TOP RIGHT: Cow A - CH4 & CH4Temp
ax2_left = axes[0, 1]
ax2_right = ax2_left.twinx()

# Plot CH4
ax2_left.plot(df_cow_A_gas['timestamp'], df_cow_A_gas['ch4'], label='CH4 (Concentration)', color='orange')
ax2_left.set_xlabel('Time')
ax2_left.set_ylabel('CH4 Concentration (%)')
ax2_left.set_title('Cow A: CH4 & CH4Temp Over Time - 13/08/2024')

# Plot CH4Temp
ax2_right.scatter(df_cow_A_gas['timestamp'], df_cow_A_gas['ch4temp'], label='CH4 Temp', 
                 color='red', s=10, alpha=0.7)
ax2_right.set_ylabel('CH4 Temperature (°C)')

# x-axis to show only HH:MM 
ax2_left.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
ax2_left.xaxis.set_major_locator(mdates.MinuteLocator(byminute=range(0, 60, 10)))

# legends
lines_left, labels_left = ax2_left.get_legend_handles_labels()
lines_right, labels_right = ax2_right.get_legend_handles_labels()
ax2_right.legend(lines_left + lines_right, labels_left + labels_right, loc='upper left')

# ======================
# MIDDLE ROW: Cow B Gas Concentration
# ======================

# MIDDLE LEFT: Cow B - CO2 & CO2Temp
ax3_left = axes[1, 0]
ax3_right = ax3_left.twinx()

# Plot CO2 (line) on the left axis
ax3_left.plot(df_cow_B_gas['timestamp'], df_cow_B_gas['co2'], label='CO2 (Concentration)', color='blue')
ax3_left.set_xlabel('Time')
ax3_left.set_ylabel('CO2 Concentration (%)')
ax3_left.set_title('Cow B: CO2 & CO2Temp Over Time - 14/08/2024')

# Plot CO2Temp (scatter) on the right axis
ax3_right.scatter(df_cow_B_gas['timestamp'], df_cow_B_gas['co2temp'], label='CO2 Temp', 
                 color='green', s=10, alpha=0.7)
ax3_right.set_ylabel('CO2 Temperature (°C)')

# Format the x-axis to show only HH:MM (no date)
ax3_left.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
ax3_left.xaxis.set_major_locator(mdates.MinuteLocator(byminute=range(0, 60, 10)))

# Combine legends
lines_left, labels_left = ax3_left.get_legend_handles_labels()
lines_right, labels_right = ax3_right.get_legend_handles_labels()
ax3_right.legend(lines_left + lines_right, labels_left + labels_right, loc='upper left')

# MIDDLE RIGHT: Cow B - CH4 & CH4Temp
ax4_left = axes[1, 1]
ax4_right = ax4_left.twinx()

# Plot CH4 (line) on the left axis
ax4_left.plot(df_cow_B_gas['timestamp'], df_cow_B_gas['ch4'], label='CH4 (Concentration)', color='orange')
ax4_left.set_xlabel('Time')
ax4_left.set_ylabel('CH4 Concentration (%)')
ax4_left.set_title('Cow B: CH4 & CH4Temp Over Time - 14/08/2024')

# Plot CH4Temp (scatter) on the right axis
ax4_right.scatter(df_cow_B_gas['timestamp'], df_cow_B_gas['ch4temp'], label='CH4 Temp', 
                 color='red', s=10, alpha=0.7)
ax4_right.set_ylabel('CH4 Temperature (°C)')

# Format the x-axis to show only HH:MM (no date)
ax4_left.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
ax4_left.xaxis.set_major_locator(mdates.MinuteLocator(byminute=range(0, 60, 10)))

# Combine legends
lines_left, labels_left = ax4_left.get_legend_handles_labels()
lines_right, labels_right = ax4_right.get_legend_handles_labels()
ax4_right.legend(lines_left + lines_right, labels_left + labels_right, loc='upper left')

# ======================
# BOTTOM ROW: Comparing Cows with Moving Averages
# ======================

# Calculate moving averages with window size 200
window_size = 200
df_cow_A_gas['co2_ma'] = df_cow_A_gas['co2'].rolling(window=window_size).mean()
df_cow_A_gas['ch4_ma'] = df_cow_A_gas['ch4'].rolling(window=window_size).mean()
df_cow_B_gas['co2_ma'] = df_cow_B_gas['co2'].rolling(window=window_size).mean()
df_cow_B_gas['ch4_ma'] = df_cow_B_gas['ch4'].rolling(window=window_size).mean()

# Create normalized time for comparing cows (minutes from start)
# This handles the different recording times
df_cow_A_gas['minutes'] = (df_cow_A_gas['timestamp'] - df_cow_A_gas['timestamp'].min()).dt.total_seconds() / 60
df_cow_B_gas['minutes'] = (df_cow_B_gas['timestamp'] - df_cow_B_gas['timestamp'].min()).dt.total_seconds() / 60

# BOTTOM LEFT: CO2 Comparison (Cow A vs Cow B)
ax5 = axes[2, 0]
ax5.plot(df_cow_A_gas['minutes'], df_cow_A_gas['co2_ma'], 'deeppink', linewidth=2.0, 
         label=f'Cow A CO2 ({window_size}-pt Moving Avg)')
ax5.plot(df_cow_B_gas['minutes'], df_cow_B_gas['co2_ma'], 'slategray', linewidth=2.0, 
         label=f'Cow B CO2 ({window_size}-pt Moving Avg)')
ax5.set_title(f'CO2 Comparison: Cow A vs Cow B (Window Size: {window_size} points)')
ax5.set_xlabel('Minutes from start')
ax5.set_ylabel('CO2 Concentration (%)')
ax5.grid(True)
ax5.legend()

# BOTTOM RIGHT: CH4 Comparison (Cow A vs Cow B)
ax6 = axes[2, 1]
ax6.plot(df_cow_A_gas['minutes'], df_cow_A_gas['ch4_ma'], 'deeppink', linewidth=2.0, 
         label=f'Cow A CH4 ({window_size}-pt Moving Avg)')
ax6.plot(df_cow_B_gas['minutes'], df_cow_B_gas['ch4_ma'], 'slategray', linewidth=2.0, 
         label=f'Cow B CH4 ({window_size}-pt Moving Avg)')
ax6.set_title(f'CH4 Comparison: Cow A vs Cow B (Window Size: {window_size} points)')
ax6.set_xlabel('Minutes from start')
ax6.set_ylabel('CH4 Concentration (%)')
ax6.grid(True)
ax6.legend()

# Format bottom row x-axis to show minutes
def minutes_formatter(x, pos):
    return f"{int(x)}"

for ax in [ax5, ax6]:
    ax.xaxis.set_major_formatter(plt.FuncFormatter(minutes_formatter))
    ax.xaxis.set_major_locator(plt.MultipleLocator(10))  # Every 10 minutes

# Adjust spacing between subplots
plt.tight_layout()
plt.show()
```
